```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
hp_dataframe = pd.read_csv("~/Desktop/hp_characters.csv")
```

```python
hp_dataframe
```

```python
hufflepuff_true = hp_dataframe.Description.str.contains("Hufflepuff")
```

```python
ravenclaw_true = hp_dataframe.Description.str.contains("Ravenclaw")
```

```python
hp_dataframe['Hufflepuff']= hufflepuff_true
```

```python
hp_dataframe['Ravenclaw']= ravenclaw_true
```

```python
slytherin_true = hp_dataframe.Description.str.contains("Slytherin")
```

```python
gryffindor_true = hp_dataframe.Description.str.contains("Gryffindor")
```

```python
hp_dataframe['Slytherin']= slytherin_true
```

```python
hp_dataframe['Gryffindor']= gryffindor_true
```

```python
chars_with_houses = hp_dataframe[hp_dataframe.Hufflepuff | hp_dataframe.Gryffindor | hp_dataframe.Slytherin | hp_dataframe.Ravenclaw]
```

```python
chars_with_houses['House'] = pd.Series(np.NaN, index=chars_with_houses.index)
```

```python
chars_with_houses
```

```python
def fill_char_house(row):
    if row["Hufflepuff"]:
        row.House="Hufflepuff"
    if row["Ravenclaw"]:
        row.House="Ravenclaw"
    if row["Slytherin"]:
        row.House="Slytherin"
    if row["Gryffindor"]:
        row.House="Gryffindor"
    return row
```

```python
chars_with_houses=chars_with_houses.apply(fill_char_house,axis=1)
```

```python
columns = ['Hufflepuff', 'Ravenclaw','Slytherin','Gryffindor']
chars_with_houses.drop(columns, inplace=True, axis=1)
```
